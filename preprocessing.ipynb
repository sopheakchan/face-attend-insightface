# Extract frame from video, using 4 frame skipping at a time

```python
from pathlib import Path
import cv2

BASE_DIR = Path.cwd()

def extract_frames(video_path, output_dir, person_name, frame_skip=4):
    video_path = Path(video_path)
    output_dir = Path(output_dir)
    output_dir.mkdir(parents=True, exist_ok=True)

    cap = cv2.VideoCapture(str(video_path))
    if not cap.isOpened():
        raise ValueError(f"Cannot open video: {video_path}")

    step = frame_skip + 1
    frame_index = 0
    saved_count = 0

    while True:
        ok, frame = cap.read()
        if not ok:
            break

        if frame_index % step == 0:
            saved_count += 1
            filename = f"{person_name}_frame{saved_count}.jpg"
            cv2.imwrite(str(output_dir / filename), frame)

        frame_index += 1

    cap.release()

    return {
        "video": str(video_path),
        "saved_frames": saved_count,
        "total_frames": frame_index,
        "frame_skip": frame_skip,
        "output_dir": str(output_dir),
    }
```

```python
videos = [
    (BASE_DIR / "vit-video.mp4", BASE_DIR / "vit", "vit"),
]

results = []
for video_path, output_dir, person_name in videos:
    info = extract_frames(
        video_path=video_path,
        output_dir=output_dir,
        person_name=person_name,
        frame_skip=4,
    )
    results.append(info)

results
```
